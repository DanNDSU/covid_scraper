```python
from bs4 import BeautifulSoup
import requests
import re
import socket 
from _thread import *
import threading 
```

```python
def crawler(url):
    try: 
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        title = soup.find_all('title')
        texts = soup.find_all('p')
        links = soup.find_all('a', attrs={'href': re.compile("^http")})

        tf_results=[]
        url_list = []
        target=[]
        keyword1='covid-19'
        keyword2='coronavirus'

        for link in links:
            address = link.get('href')
            try: 
                #print(address)
                page2 = requests.get(address)
                soup2 = BeautifulSoup(page2.text, 'html.parser')
                title2 = soup2.find_all('title')
                texts2 = soup2.find_all('p')
                links2 = soup2.find_all('a', attrs={'href': re.compile("^http")}) 
                for idx in range(len(texts2)):
                    L = texts2[idx].get_text()
                    tf_value = (keyword1.lower() in L.lower()) or (keyword2.lower() in L.lower())
                    tf_results.append(tf_value)
                    if tf_value == True:
                        target.append(L)
                    else:
                        pass

            except Exception as ie:
                pass

            if any(True == s for s in tf_results):
                url_list.append([address, target]) 
        
        
    except Exception as ie:
        pass
            
    return url_list
```

```python
def crawler_repeat(url):
    first_crawler = crawler(url)
    print('first level url crawled !')
    #collect the second level urls
    second_lev_urls = []
    for item in l1:
        second_lev_urls.append(item[0])
    
    n1=len(second_lev_urls)
    n2=5
    for i in range(n2):
        second_lev_result = crawler(second_lev_urls[i])
        l1[i].append(second_lev_result[:10])
        print('seconde level url ', i , ' crawled !')
    
    return first_crawler
```
