```python
from bs4 import BeautifulSoup
import requests
import re
import socket 
from _thread import *
import threading 
```

```python
url='https://www.health.nd.gov/diseases-conditions/coronavirus/north-dakota-coronavirus-cases'
#url= 'https://www.cdc.gov/coronavirus/2019-ncov/cases-updates/cases-in-us.html'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
# prints any titles from the first page hit.
title = soup.find_all('title')
texts = soup.find_all('p')
links = soup.find_all('a', attrs={'href': re.compile("^http")})
```

```python
tf_results=[]
results = []
url_list = []
target=[]
keyword1='covid-19'
keyword2='coronavirus'

for link in links:
    address = link.get('href')
    try: 
        print(address)
        page2 = requests.get(address)
        soup2 = BeautifulSoup(page2.text, 'html.parser')
        title2 = soup2.find_all('title')
        texts2 = soup2.find_all('p')
        links2 = soup2.find_all('a', attrs={'href': re.compile("^http")}) 
        # loop through texts on address html, compare with keywords, if true, return this address and related texts
        for idx in range(len(texts2)):
            L = texts2[idx].get_text()
            tf_value = (keyword1.lower() in L.lower()) or (keyword2.lower() in L.lower())
            tf_results.append(tf_value)
            if tf_value == True:
                target.append(L)
                print(target)
            else:
                #print('None')
                pass
              
    except Exception as ie:
        pass
    
    if any(True == s for s in tf_results):
        url_list.append([address, target])   
        # along with this address, any information can be sent back to the client, tile, related texts, or links on this html. 
```
